```python
#clean data set
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    tokens = word_tokenize(text)  # Tokenize the text
    tokens = [word for word in tokens if word.isalpha()]  # Remove non-alphabetic tokens
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    return ' '.join(tokens)
```

```python
data1['cleaned_text'] =data1['Tweet'].apply(clean_text)

data1['Stance'] = np.random.randint(2, size=len(data1))
```

```python
#using word embeding Word2vec

all_words = data1['cleaned_text'].tolist()
word2vec_model = Word2Vec(all_words, vector_size=100, window=5, min_count=1, workers=4)

def get_avg_word2vec(tokens, model, vector_size):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(vector_size)
        vector_size =100

```

```python
# Using bag of words instead of word2vec for word embedding
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data1['cleaned_text'])
```

```python

data1['tweet_embedding'] = data1['cleaned_text'].apply(lambda text: get_avg_word2vec(text.split(), word2vec_model, vector_size))
```

```python
X = np.array(data1['tweet_embedding'].tolist())
y = data1['Stance']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)
```
